```python
df=pd.read_json('https://api.steampowered.com/ISteamApps/GetAppList/v0002/')
```

```python
df=pd.json_normalize(df['applist'].iloc[0])
```

```python
df.head()
```

```python
df.rename(columns={'appid':'item_id'}, inplace=True)
```

```python
dfnames=pd.merge(ids,df,on='item_id',how='inner')
```

```python
dfnames.head()
```

```python
i=315392
dfnames[dfnames['item_id']==i]['name'].iloc[0]
```

```python
li=list(dfnames['item_id'])
```

```python
li
```

```python
def get_name(i):
  if i in li:
    name=dfnames[dfnames['item_id']==i]['name'].iloc[0]
  else:
    name=f'id: {i} sin nombre'
  return name
```

```python
get_name(1840)
```

```python
ids['item_name']=ids['item_id'].apply(get_name)
```
