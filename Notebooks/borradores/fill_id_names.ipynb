```python
import pandas as pd
```

```python
games = pd.read_parquet('../Datasets/gamesv5.parquet')
```

```python
reviews = pd.read_parquet('../Datasets/reviews.parquet')
```

```python
games = games.drop_duplicates('item_id')
```

```python
reviews = reviews.drop_duplicates('item_id')
```

```python
reviews.info()
```

```python
games.info()
```

```python
# Supongamos que 'item_id' es la columna que se utiliza para la coincidencia
merged_df = pd.merge(games, reviews, on='item_id', how='left', indicator=True)

# Filtrar solo las filas que no tienen coincidencias en 'games'
reviews_sin_coincidencias = merged_df[merged_df['_merge'] == 'left_only']

# Eliminar la columna de indicador '_merge' si es necesario
reviews_sin_coincidencias = reviews_sin_coincidencias.drop('_merge', axis=1)

# Ahora 'reviews_sin_coincidencias' contiene las filas de 'reviews' sin coincidencias en 'games'
```

```python
fid = reviews_sin_coincidencias[['item_id','release_date']].copy()
```

```python
df=pd.read_json('https://api.steampowered.com/ISteamApps/GetAppList/v0002/')
```

```python
df=pd.json_normalize(df['applist'].iloc[0])
```

```python
df.rename(columns={'appid':'item_id'}, inplace=True)
```

```python
li=list(df['item_id'])
```

```python
li
```

```python
def get_name(i):
  if i in li:
    name=df[df['item_id']==i]['name'].iloc[0]
  else:
    name=f'id: {i} sin nombre'
  return name
```

```python
fid['item_name']=fid['item_id'].apply(get_name)
```
