```python
import pandas as pd # Para trabajar con dataframes
import numpy as np # Para operaciones y funciones con arrays
```

```python
games = pd.read_json("../Datasets/steam_games.json.gz", lines=True)
```

```python
games.head()
```

```python
games.columns
```

```python
games.drop(columns=['publisher', 'genres', 'app_name', 'url', 'reviews_url', 'specs', 'price', 'early_access', 'developer'], inplace=True)
```

```python
games.info()
```

```python
games.dropna(subset='id', inplace=True)
```

```python
games.dropna(subset='release_date', inplace=True)
```

```python
games.dropna(subset='tags', inplace=True)
```

```python
games.drop_duplicates(subset='id', inplace=True)
```

```python
games.info()
```

```python
games['release_date']
```

```python
games['release_date'] = games['release_date'].str.extract(r'(\d{4})')
```

```python
games=games[~games['release_date'].isna()]
```

```python
games['release_date'] = games['release_date'].astype(int)
```

```python
games['release_date'].min()
```

```python
games['release_date'].max()
```

```python
games[games['release_date']>2023]
```

```python
games=games[games['release_date']<2023]
```

```python
games.info()
```

```python
games.rename(columns={'tags': 'genres'}, inplace=True)
```

```python
games.rename(columns={'id': 'item_id'}, inplace=True)
```

```python
games.head()
```

```python
games.info()
```

```python
games['item_id'] = games['item_id'].astype(int)
```

```python
games=games.sort_values('item_id')
```

```python
games=games.explode('genres')
```

```python
games.head()
```

```python
dummies = pd.get_dummies(games['genres'], dtype=np.int8)
dummies
```

```python
dummies.sum()>dummies.mean()
```

```python
dummies.loc[:, dummies.sum() > dummies.sum().mean()]
```

```python
col=dummies.loc[:, dummies.sum() > dummies.sum().mean()].columns
```

```python
col=col.to_list()
```

```python
dummies=dummies[col]
```

```python
dummies
```

```python
dummies.columns=dummies.columns.str.upper()
```

```python
dummies
```

```python
games.head()
```

```python
df = pd.concat([games, dummies], axis=1)
```

```python
df.info()
```

```python
df.drop(columns=['title','release_date','genres'], inplace=True)
```

```python
df
```

```python
col=df.columns.to_list()
```

```python
df=df.groupby('item_id')[col].sum()
```

```python
df
```

```python
df.drop(columns=['item_id'], inplace=True)
```

```python
df=df.reset_index()
```

```python
df
```

```python
df.to_parquet("../Datasets/gamesoh.parquet")
```
