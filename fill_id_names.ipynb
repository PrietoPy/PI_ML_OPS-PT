```python
def get_name(id):
  url = f"https://api.steampowered.com/ISteamNews/GetNewsForApp/v0002/?appid={id}&count=3&maxlength=300&format=json"
  jst = pd.read_json(url, orient='records')
  appnews_data = jst['appnews'].iloc[2]
  if len(appnews_data) == 0:
    return 'Id: '+str(id)+' no name'
  else:
    return appnews_data[0]['title']
```

```python
get_name(315396)
```

```python
for ind, row in ids.iterrows():
  ids.at[ind, 'item_name'] = get_name(row['item_id'])
```
